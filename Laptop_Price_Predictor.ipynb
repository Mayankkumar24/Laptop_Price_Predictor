```python
def Extract_Generation(x):
    mp = {
        '1st' : '1st',
        '2nd' : '2nd',
        '3rd' : '3rd',
        '4th' : '4th',
        '5th' : '5th',
        '6th' : '6th',
        '7th' : '7th',
        '8th' : '8th',
        '9th' : '9th',
        '10th': '10th',
        '11th': '11th',
        '12th': '12th',
        '13th': '13th',
        '14th': '14th',
        '15th': '15th'
    }
    gen_ = " ".join(x.split()[0:1])
    if (gen_ in mp):
        return mp[gen_]
    else:
        return gen_
laptop_data['Generation'] = laptop_data['processor'].apply(lambda x : Extract_Generation(x))
```

```python
def Extract_Processor(x):
    if ('Intel' in x):
        if ('i3' in x):
            return 'Intel i3'
        elif ('i5' in x):
            return 'Intel i5'
        elif ('i7' in x):
            return 'Intel i7'
        elif ('i9' in x):
            return 'Intel i9'
        else:
            return 'Other Intel Processor'
    elif ('AMD' in x):
        if ('5' in x):
            return 'AMD Ryzen 5'
        elif ('7' in x):
            return 'AMD Ryzen 7'
        else:
            return 'Other AMD Processor'
    elif ('Apple' in x):
        return 'Apple Processor'
    else:
        return 'Other Processor'
laptop_data['Processor'] = laptop_data['processor'].apply(lambda x : Extract_Processor(x))
```

```python
laptop_data.drop('processor',axis =1, inplace = True)
```

```python
laptop_data.tail()
```

```python
replacement = {'2GB':2,'4GB':4,'8GB':8,'12GB':12,'16GB':16,'32GB':32,'64GB':64}
laptop_data['Ram'].replace(replacement,inplace=True)
laptop_data['ROM'].unique()
```

```python
def Extract_Rom(x):
    if ('GB' in x):
        if (len(x)==5):
            return int(x[0:3])
        else:
            return int(x[0:2])
    else:
        return 1024*(int(x[0:1]))
laptop_data['NewROM'] = laptop_data['ROM'].apply(lambda x : Extract_Rom(x))
```

```python
laptop_data.drop('ROM',axis=1,inplace=True)
```

```python
def Extract_GPU(x):
    if ('AMD' in x):
        return 'AMD'
    elif ('Intel' in x):
        return 'Intel'
    elif ('Apple' in x):
        return 'Apple'
    elif ('NVIDIA' in x):
        return 'NVIDIA'
    elif ('ARM' in x):
        return 'ARM'
    else:
        return 'Other'
laptop_data['NewGPU'] = laptop_data['GPU'].apply(lambda x: Extract_GPU(x))
laptop_data.drop('GPU',axis=1,inplace=True)
```

```python
X = laptop_data.drop(columns=['price'])
Y = np.log(laptop_data['price'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.15,random_state=42)
```

```python
laptop_data.drop('price',axis=1,inplace=True)
```

```python
categorical_transformer1 = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
numeric_transformer1 = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])
preprocessor1 = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer1, [0, 2, 3, 7, 9, 10, 12]),
        ('num', numeric_transformer1, [1, 4, 5, 6, 8, 11])
    ]
)
lr = LinearRegression()
pipeline1 = Pipeline([
    ('preprocessor', preprocessor1),
    ('lr', lr)
])
pipeline1.fit(X_train, Y_train)
pred1 = pipeline1.predict(X_test)
print("R2 Score :",r2_score(Y_test,pred1))
print("MAE      :",mean_absolute_error(Y_test,pred1))
```

```python
categorical_transformer2 = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='constant',fill_value='missing')),
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])
numeric_transformer2 = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])
preprocessor2 = ColumnTransformer(
            transformers=[('cat',categorical_transformer2,[0,2,3,7,9,10,12]),
                         ('num',numeric_transformer2,[1,4,5,6,8,11])])
rf = RandomForestRegressor()
pipeline2 = Pipeline([('preprocessor',preprocessor2),
                     ('rf',rf)])
pipeline2.fit(X_train,Y_train)
pred2 = pipeline2.predict(X_test)
print("R2 Score :",r2_score(Y_test,pred2))
print("MAE      :",mean_absolute_error(Y_test,pred2))
pipeline2
```

```python
categorical_transformer3 = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='constant',fill_value='missing')),
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])
numeric_transformer3 = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])
preprocessor3 = ColumnTransformer(
            transformers=[('cat',categorical_transformer3,[0,2,3,7,9,10,12]),
                         ('num',numeric_transformer3,[1,4,5,6,8,11])])
dt = DecisionTreeRegressor()
pipeline3 = Pipeline([('preprocessor',preprocessor3),
                     ('dt',dt)])
pipeline3.fit(X_train,Y_train)
pred3 = pipeline3.predict(X_test)
print("R2 Score :",r2_score(Y_test,pred3))
print("MAE      :",mean_absolute_error(Y_test,pred3))
```
